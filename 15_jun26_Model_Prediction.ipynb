```python
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import tensorflow as tf
```

```python
with open('15jun26_label_encoder.pkl','rb') as f:
    le=pickle.load(f)
```

```python
with open('15jun26_one_hot_encoder.pkl','rb') as f:
    onehot=pickle.load(f) 
```

```python
with open('15jun26_Scaler.pkl','rb') as f:
    scaler=pickle.load(f)
```

```python
model=load_model('15jun26_ANNmodel.h5')
```

```python
user={'Pclass':3,'Sex':'female','SibSp':1,'Parch':1,'Fare':12.25,'Embarked':'Q'}
user
```

# converting the data into dataframe format 

```python
user=pd.DataFrame([user])
```

```python
user
```

# how to make use of label encoder and one hot encoder to transform the user data

```python
user['Sex']=le.transform(user['Sex']) 
# we will not use fit_tranform- this will be used when we are training the model.
#when we are making prediction we will use transform only. because we have already fitted the label encoder on the training data. so we will use the same label encoder to transform the user data.
```

```python
embarked=onehot.transform(user[['Embarked']]).toarray()
embarked
```

```python
Embarked=pd.DataFrame(embarked,columns=onehot.get_feature_names_out())
Embarked
```

```python
user=pd.concat([user.drop(columns=['Embarked']),Embarked],axis=1)
```

# Scaling the user data

```python
num_cols=['Pclass','SibSp','Parch','Fare']
```

```python
user[num_cols]=scaler.transform(user[num_cols])
```

```python
user
```

# Prediction

```python
y=model.predict(user)
```

```python
y
```

# The out is the probability of the user survived or not. if the value is greater than 
# 0.5 then we can say that the user is survived otherwise not survived.

```python
Y=y[0][0]
```

```python
if y>0.5 :
    print('The person is likely to survive')
else:
    print('The person is likely to not survive')    
```
